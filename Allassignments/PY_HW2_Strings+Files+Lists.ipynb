# Validating Username and Password

```python
def validateUserPass(inUser, inPass):
    retValue = True
    if not inUser.isalpha() :
        print("Username must contain just Alphabetic characters")
        retValue = False
    if 3 > len(inUser) or len(inPass) > 8 :
        print("Username length must be at least 3 and as most 8 characters")
        retValue = False
    if 3 > len(inPass) or len(inPass) > 6 :
        print("Password length must be at least 3 and as most 6 characters")
        retValue = False
    if not inPass.isalnum() :
        print("Password can contains just AlphaNumeric characters")
        retValue =  False
    return retValue
```

# Reading the whole content of the user's text file

```python
def read_text_file():
    file_name = input('Input file name:')
    try:
        file = open(file_name, 'r')
        print(f"Text file '{file_name}' read successfully.")
        return file.read().lower()
    except:
        print(f"Error: File '{file_name}' not found.")
        return None
```

# Splitting the text into a list of words and printing the length of the text 

```python
def word_count(text):
    words = text.split()
    print(f"Total number of words in the text: {len(words)}")
    return words
```

# Finding the longest word

```python
def longest_word(wordsList):
    sumLen = 0
    maxLen = 0
    maxWord = ''
    for w in wordsList:
        sumLen += len(w)
        if maxLen < len(w):
            maxLen = len(w)
            maxWord = w
    msg = f"Longest Word: '{maxWord}', Length: {maxLen}\n"
    avgLen = sumLen / len(wordsList)
    msg += f"Average Words Length: {avgLen}\n"
    print(msg)
```

# Cleaning all-num words form the list of words

```python
def clean_words(wordsList):
    newList = list()
    for w in wordsList:
        if w.isnumeric():
            continue
        newList.append(w)
    print(f"Total number of words after cleaning: {len(newList)}")
    return newList
```

# Filtering words that are in the balcklist

```python
def filter_words(wordsList):
    blackfile = open('blacklist.txt','r')
    blackList = blackfile.read().lower().split()
    newList = wordsList.copy()
    for w in blackList:
        while w in newList:
            newList.remove(w)
    print(f"Total number of words after filtering: {len(newList)}")
    return newList
```

# Computing words frequencies

```python
def word_frequency(wordsList):
    uniqWords = list()
    wordsFreq = list()
    for w in wordsList:
        if w in uniqWords:
            ind = uniqWords.index(w)
            wordsFreq[ind] += 1
        else:
            uniqWords.append(w)
            wordsFreq.append(1)
            
    maxFreq = max(wordsFreq)
    maxWord = uniqWords[wordsFreq.index(maxFreq)]
    avgFreq = sum(wordsFreq) / len(wordsFreq)
    msg = f"Number of Unique Words: {len(uniqWords)}\nMost Frequent Word: '{maxWord}'-> frequency: {maxFreq}\nAverageFrequency is: {avgFreq}\n"
    print(msg)
```

# Writing Cleaned and Filtered text to a new file

```python
def writeFilteredText(orgText):
    sentenceList = orgText.split('\n')
    newText = ''
    for sentence in sentenceList :
        if sentence != '' :
            newText += ' '.join(filter_words(clean_words(word_count(sentence)))) + '\n'
    outFile = open ("filteredText.txt",'w')
    outFile.write(newText)
    outFile.close()
    return
```

# Making  text processing menu

```python
def userPanel():
    text = None
    while True:
        print("\n1. Read Text File\n2. Word Count\n3. Longest Word\n4. Clean Words\n5. Filter Words\n6. Frequency of Each Word\n7. Write down filtered text\n8. Logout")
        choice = input("Enter your choice (1-8): ")

        if choice == '1':
            text = read_text_file()

        elif choice == '2' and text is not None:
            rawWords = word_count(text)

        elif choice == '3' and text is not None:
            longest_word(rawWords)

        elif choice == '4' and text is not None:
            cleanedWords = clean_words(rawWords)

        elif choice == '5' and text is not None:
            filteredWords = filter_words(cleanedWords)

        elif choice == '6' and text is not None:
            word_frequency(filteredWords)

        elif choice == '7' and text is not None:
            writeFilteredText(text)
        
        elif choice == '8':
            print("Thank you for using my Text Preprocessing module.\n You've been logged out.")
            break
            
        elif text is None:
            print("The input file must be read at first.")

        else:
            print("Invalid choice. Please enter a number between 1 and 8.")
    
    return
```

# Providing main menu and managing register/login

```python
def mainMenu() :
    while True:
        try:
            userFile = open("userInfo.txt","r")
            print("\n1. Login")
            print("2. Exit")
            cmd = input("Enter Your Choice (1 or 2):")
            try:
                cmd = int(cmd)
            except:
                print("Invalid Choice! you must enter 1 or 2")
                continue
                
            if cmd == 1 :
                username = input("Enter a Username:")
                password = input("Enter a Password:")
                lines = userFile.readlines()
                if username == lines[0].rstrip() and password == lines[1].rstrip():
                    userPanel()
                else:
                    print("ERROR: Wrong Username or Password! try again...")
            elif cmd == 2:
                print("Goodbye...")
                break
            else:
                print("Invalid Choice! you must enter 1 or 2")
        except:
            print("\n1. Register")
            print("2. Exit")
            cmd = input("Enter Your Choice (1 or 2):")
            try:
                cmd = int(cmd)
            except:
                print("Invalid Choice! you must enter 1 or 2")
                continue
                
            if cmd == 1 :
                username = input("Enter a Username:")
                password = input("Enter a Password:")
                if (validateUserPass(username, password)) :
                    userFile = open("userInfo.txt","w")
                    userFile.write(f"{username}\n{password}")
                    userFile.close()
            elif cmd == 2:
                print("Goodbye...")
                break
            else:
                print("Invalid Choice! you must enter 1 or 2")
    return
```
